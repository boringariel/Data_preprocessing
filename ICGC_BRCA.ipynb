```python
import pandas as pd
pd.set_option('display.max_columns', 600)
```

```python
brca2 = pd.read_csv('./HRT_O/ICGC_BRCA-EU/donor_therapy.BRCA-EU.tsv', sep='\t')
brca3 = pd.read_csv('./HRT_O/ICGC_BRCA-EU/donor.BRCA-EU.tsv', sep='\t')
brca4 = pd.read_csv('./HRT_O/ICGC_BRCA-EU/simple_somatic_mutation.open.BRCA-EU.tsv', sep='\t')
```
